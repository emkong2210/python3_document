```python
# threadex1.py
import threading, time

# 스레드로 실행할 함수를 정의한다.
def myThread(id):
    for i in range(10):
        print('id={} --> {}'.format(id, i))
        # CPU를 양도한다.
        time.sleep(0)

# 스레드 객체를 모아 두는 리스트이다.        
threads = []
# 스레드 두 개를 생성한다.
for i in range(2):
    th = threading.Thread(target = myThread, args = (i,))
    # 스레드를 실행한다.
    th.start()
    # 스레드 객체를 리스트에 저장한다.
    threads.append(th)
    
for th in threads:
    # 각 스레드가 종료할 때까지 대기한다.
    th.join()
    
print('Exiting')
```

```python
# threadex2.py
import threading, time

# Thread 클래스의 하위 클래스를 정의한다.
class MyThread(threading.Thread):
    # run() 메서드는 스레드가 실행할 코드를 갖는다.
    def run(self):
        for i in range(10):
            print('id={} --> {}'.format(self.getName(), i))
            # CPU를 양도한다.
            time.sleep(0)
            
threads = []

for i in range(2):
    # 스레드 객체를 얻고
    th = MyThread()
    # 스레드를 실행한다.
    th.start()
    # 스레드 종료를 검사하기 위해 리스트에 잠시 저장한다.
    threads.append(th)
    
for th in threads:
    # 스레드 th가 종료할 때까지 대기한다.
    th.join()
    
print('Exiting')
```

```python
# threadex4.py
import threading, time
import random

# 소비자 프로세스의 수
NR_CONSUMER = 10
# 생산자 프로세스의 수
NR_PRODUCER = NR_CONSUMER // 2

# 공유 버퍼
buffer = []
# 조건 변수 생성
cv = threading.Condition()

# 소비자
class Consumer(threading.Thread):
    def run(self):
        # 한 소비자는 5개의 정보를 소비한다.
        for x in range(5):
            cv.acquire()
            # 정보가 없으면
            while len(buffer) < 1:
                print('waiting..')
                # 기다린다.
                cv.wait()
            # 정보를 소비한다.
            print(buffer.pop(0))
            cv.release()
            # 잠시 대기한다.
            time.sleep(0.01)

# 생산자
class Producer(threading.Thread):
    def run(self):
        # 한 생성자는 10개의 정보를 생산한다.
        for x in range(10):
                cv.acquire()
                buffer.append(random.randrange(0, 20))
                # cv 대기 큐에서 대기하는 스레드가 있으면 깨운다
                cv.notify()
                cv.release()
                # 잠시 대기한다.
                time.sleep(0.02)

# 소비자 스레드 리스트                
con = []
# 생산자 스레드 리스트
pro = []

for i in range(NR_CONSUMER):
    # 소비자 스레드 생성
    con.append(Consumer())
    
for i in range(NR_PRODUCER):
    # 생산자 스레드 생성
    pro.append(Producer())
    
for th in con:
    # 소비자 스레드를 먼저 시작한다.
    th.start()
    
for th in pro:
    # 생산자 스레드를 시작한다.
    th.start()
    
for th in con:
    # 소비자 스레드가 종료할 떄까지 대기한다.
    th.join()
    
for th in pro:
    # 생산자 스레드가 종료할 떄까지 대기한다.
    th.join()
    
print('Exiting')
```

```python
# threadex5.py
import threading

# 세마포어 객체 생성
sem = threading.Semaphore(3)

class RestrictedArea(threading.Thread):
    def run(self):
        for x in range(500):
            # 필요한 처리가 있으면 한다.
            sem.acquire()
            print('id={} --> {}'.format(self.getName(), sem._value))
            # 임계 영역 안에서 수행할 작업
            sem.release()
            # 필요한 처리가 있으면 한다.

thlist = []

# 100개의 스레드 객체 생성
for i in range(100):
    thlist.append(RestrictedArea())
    
for th in thlist:
    # 스레드를 시작한다.
    th.start()
    
for th in thlist:
    # 스레드가 종료할 떄까지 대기한다.
    th.join()
    
print('Exiting')
```

```python
# threadex6.py
import threading

# 이벤트 객체 생성
eve = threading.Event()

class PrepareThread(threading.Thread):
    def run(self):
        # 뭔가를 준비하고, 준비되었음을 알린다.
        eve.set()
        print('Ready')
        
class ActionThread(threading.Thread):
    def run(self):
        # 앞에서 처리할 코드가 있다면 추가한다.
        print(self.getName(), 'waiting..')
        # PrepareThread 객체가 준비를 마칠 때가지 대기한다.
        eve.wait()
        # 본격적인 작업은 여기에 기술한다.
        print(self.getName(), 'done')
        
thlist = []

for i in range(5):
    thlist.append(ActionThread())
    
for th in thlist:
    th.start()
    
PrepareThread().start()

for th in thlist:
    th.join()
    
print('Exiting')
```
